```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data_final.csv')
```

```python
data = data.drop('Unnamed: 0', axis=1)
```

```python
data.head()
```

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

# Linear Discriminant Analysis(LDA)

```python
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import MinMaxScaler

random_state = []

for j in range(1,101): 
    random_state.append(j)
ai = []
for rand in random_state:
  X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state= rand,stratify=y)
  log_reg_kf = LinearDiscriminantAnalysis()
  log_reg_kf.fit(X_train, y_train)
  pred_proba_1 = log_reg_kf.predict_proba(X_test)[:,1]
  auc_test = roc_auc_score(y_test, pred_proba_1)
  print('random_state=', rand,'  ' ,'auc_test =', auc_test)
  print('')
  ai.append(auc_test)
print("평균 = ", np.mean(ai))
```

# SVM

```python
from sklearn.svm import SVC
model = SVC() 


from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import MinMaxScaler

random_state = []

for j in range(1,101): 
    random_state.append(j)
ai = []
for rand in random_state:
  X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state= rand,stratify=y)
  model = SVC(probability=True) 
  model.fit(X_train, y_train)
  pred_proba_1 = model.predict_proba(X_test)[:,1]
  auc_test = roc_auc_score(y_test, pred_proba_1)
  print('random_state=', rand,'  ' ,'auc_test =', auc_test)
  print('')
  ai.append(auc_test)
print("평균 = ", np.mean(ai))
```

# QDA(Quadratic Discriminant Analysis)

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import MinMaxScaler

random_state = []

for j in range(1,101): 
    random_state.append(j)
ai = []
for rand in random_state:
  X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state= rand,stratify=y)
  qda = QuadraticDiscriminantAnalysis(store_covariance=True)
  qda.fit(X_train, y_train)
  pred_proba_1 = qda.predict_proba(X_test)[:,1]
  auc_test = roc_auc_score(y_test, pred_proba_1)
  print('random_state=', rand,'  ' ,'auc_test =', auc_test)
  print('')
  ai.append(auc_test)
print("평균 = ", np.mean(ai))
```
